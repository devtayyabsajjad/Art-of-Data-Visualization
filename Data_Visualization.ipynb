# 1. Introdunction

## 1.1 Overview of Python Data Visualization Tools

Python has three main tools for **drawing graphs and charts**:

### 1.1.1 matplotlib

- **Matplotlib** is a popular Python library for creating **graphs and charts**.  
- It is used to draw **line plots**, **bar charts**, **scatter plots**, **histograms**, and more.  
- It is **flexible** and allows for **customizing** every part of a plot.  
- **Seaborn** includes a **collection of sample datasets** for quick testing and learning.
- **Example:** It is best for making **simple** and **detailed** visualizations.

### 1.1.2 pyplot (Module of Matplotlib)

- It provides **better styling** and **simpler syntax** for complex visualizations.  
- While Seaborn graphs are **not fully interactive**, they are **highly customizable** and work well for **data analysis**.  
- **Example:** Use Seaborn for **heatmaps**, **box plots**, **violin plots**, and **pair plots**.

- **Plotly** is a **Python library** for creating **interactive** and **high-quality** visualizations.  
- It supports a wide range of **charts** like **line charts**, **scatter plots**, **bar charts**, **3D plots**, and **maps**.  
- **Interactive features** like **zooming**, **hovering**, and **panning** make it ideal for **exploratory data analysis**.
- **Highly customizable** with options for **themes**, **colors**, **axes**, and **annotations**.

# 2. Types of Data



Data is divided into two main categories: **Quantitative Data** (numerical) and **Qualitative Data** (categorical). Each type provides different kinds of information and serves specific analytical purposes.

---

## 2.1 Quantitative Data

**Quantitative data** represents **numerical** values that can be **measured** or **counted**. It answers questions like **"How much?"** or **"How many?"**

### ðŸ”¢ **Types of Quantitative Data:**
1. **Continuous Data**  
2. **Discrete Data**



### 2.1.1 Continuous Data

- **Definition:** Continuous data can take **any value** within a range, including **decimals** or **fractions**.  
- It is **measurable** and represents information that can have an **infinite** number of values.  

- **Examples:**  
  - Height: 5.8 feet  
  - Weight: 70.5 kg  
  - Temperature: 36.7Â°C  
  - Time: 3.45 hours  

**When to Use:**  
Use continuous data when values can take **infinite** possibilities within a given range.



### 2.1.2 Discrete Data

- **Definition:** Discrete data includes **fixed**, **countable** values that cannot be **broken down** into smaller parts.  
- It represents **whole numbers** or **specific counts**.  

- **Examples:**  
  - Number of students in a class: 30  
  - Cars in a parking lot: 12  
  - Test scores: 95  
  - Number of pets: 3  

**When to Use:**  
Use discrete data when the data consists of **separate**, **whole values** that cannot be subdivided.

---

## 2.2 Qualitative Data

**Qualitative data** (or **categorical** data) describes **categories** or **labels** rather than numbers. It answers questions like **"What kind?"** or **"Which category?"**

### ðŸ”¢ **Types of Qualitative Data:**
1. **Nominal Data**  
2. **Ordinal Data**

---

### 2.2.1 Nominal Data

- **Definition:** Nominal data represents **categories** or **labels** without any **natural order** or ranking.  
- It is **descriptive** and cannot be arranged in a logical sequence.  

- **Examples:**  
  - Gender: Male, Female  
  - Colors: Red, Blue, Green  
  - Blood Group: A, B, AB, O  
  - Types of Fruit: Apple, Banana, Mango  

**When to Use:**  
Use nominal data when the categories are **distinct** and **cannot** be logically ordered.



### 2.2.2 Ordinal Data

- **Definition:** Ordinal data represents **categories** with a **specific order** or ranking.  
- The order is **meaningful**, but the **difference** between categories is **not precisely defined**.  

- **Examples:**  
  - Education Level: High School, Bachelor, Master, PhD  
  - Customer Satisfaction: Poor, Fair, Good, Excellent  
  - Clothing Size: Small, Medium, Large, Extra Large  
  - Pain Level: Mild, Moderate, Severe  

**When to Use:**  
Use ordinal data when the **order** of the categories **matters**, but the **exact difference** is not measurable.

5. Matplotlib Object Hierarchy


In **Matplotlib**, a plot is made up of different **objects**. These objects are like building blocks that help you create and customize your plot.

At the **top** is the **Figure** (the entire plot area), and inside it, there are **Axes**, **Axis**, and other parts like **tick marks**, **lines**, **legends**, **titles**, and **text boxes**.

---

## 5.1 Main Objects in Matplotlib

1. **Figure** â€“ The **whole** plot area (can hold one or more plots).  
2. **Axes** â€“ The **individual plot** inside the figure.  
3. **Axis** â€“ The **X-axis** and **Y-axis** in a plot.  
4. **Other Elements** â€“ Extra parts inside the plot:
   - **Tick Marks** â€“ Small lines on the axis to show **scale**.  
   - **Lines** â€“ Curves or markers to **show data**.  
   - **Legend** â€“ Explains what each **line** or **marker** means.  
   - **Title** â€“ The **heading** of the plot.  
   - **Text Boxes** â€“ Custom **labels** or **annotations**.  

---




## Example

````python
import matplotlib.pyplot as plt

# Data for the plot
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 35]
y2 = [15, 25, 30, 35, 40]

# Create a figure and axes
fig, ax = plt.subplots()

# Plotting two lines
ax.plot(x, y1, label="Line 1", color="blue", linestyle="--", marker="o")  # Line 1
ax.plot(x, y2, label="Line 2", color="green", linestyle="-", marker="s")  # Line 2

# Add parts of the plot
ax.set_title("Parts of a Plot Example")      # Title
ax.set_xlabel("X-axis (Time)")               # X-axis label
ax.set_ylabel("Y-axis (Value)")              # Y-axis label
ax.legend()                                  # Legend
ax.grid(True)                                # Grid
ax.set_xticks([1, 2, 3, 4, 5])               # Custom tick marks on X-axis
ax.set_yticks([10, 20, 30, 40])              # Custom tick marks on Y-axis
ax.text(3, 28, "Important Point", color="red") # Annotation (text box)

# Display the plot
plt.show()
````

# 10. Plot Types in Matplotlib

## 10.1 Line Plot

- **Purpose:** Used to show **trends** and **changes** over time.  
- **Best For:** Visualizing **continuous** data.

### Example for Line Plot

````python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Line plot
plt.plot(x, y, marker='o', linestyle='-', color='blue')

plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
````

## 10.2 Scatter Plot
- Purpose: Displays individual points to show relationships between variables.
- Best For: Identifying patterns, clusters, and outliers.

### Example for Scatter Plot

````python
import matplotlib.pyplot as plt

# Data
x = [5, 7, 8, 10, 12]
y = [15, 18, 22, 30, 40]

# Scatter plot
plt.scatter(x, y, color='red', marker='x')

plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
````

## 10.3 Histogram
- Purpose: Shows the distribution of numerical data.
- Best For: Analyzing frequency or data ranges.

### Example for Histogram

````python
import matplotlib.pyplot as plt

# Data
data = [10, 15, 20, 25, 30, 35, 40, 30, 25, 20]

# Histogram
plt.hist(data, bins=5, color='green', edgecolor='black')

plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
````

## 10.4 Bar Chart
- Purpose: Represents categories with rectangular bars.
- Best For: Comparing discrete data or categories.

### Example for Bar Chart

````python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 30, 45]

# Bar chart
plt.bar(categories, values, color='purple')

plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
````

## 10.5 Horizontal Bar Chart
- Purpose: Similar to a bar chart, but horizontal.
- Best For: Comparing categories with long labels.

### Example for Horizontal bar Chart

````python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 30, 45]

# Horizontal bar chart
plt.barh(categories, values, color='orange')

plt.title("Horizontal Bar Chart Example")
plt.xlabel("Values")
plt.ylabel("Categories")

plt.show()
````

## 10.6 Error Bar Chart
- Purpose: Shows variability or uncertainty in the data.
- Best For: Visualizing error margins or confidence intervals.

### Example for Error Bar Chart

````python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
errors = [1, 2, 1, 3]

# Error bar chart
plt.errorbar(x, y, yerr=errors, fmt='o', color='blue')

plt.title("Error Bar Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
````

## 10.7 Stacked Bar Chart
- Purpose: Displays multiple datasets on top of each other.
- Best For: Comparing the contribution of parts to a whole.

### Example for Stacked Bar Chart

````python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C']
group1 = [5, 10, 15]
group2 = [3, 7, 9]

# Stacked bar chart
plt.bar(categories, group1, color='cyan', label='Group 1')
plt.bar(categories, group2, bottom=group1, color='magenta', label='Group 2')

plt.title("Stacked Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()

plt.show()
````

## 10.10 Area Chart

An **area chart** is a **filled line plot** that shows **cumulative data** over time or a continuous range.

### Example for Area Chart

````python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

# Area chart
plt.fill_between(x, y, color='skyblue', alpha=0.7)

# Add title and labels
plt.title("Area Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
````

# 10.11 Contour Plot  
A **contour plot** is used to **visualize 3D data** in a **2D** format by drawing **contours** (lines of equal value).

### Example for Counter Plot

````python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 5, 50)  # X-axis range
y = np.linspace(0, 5, 40)  # Y-axis range
X, Y = np.meshgrid(x, y)   # Create grid
Z = np.sin(X) * np.cos(Y)  # Z values (height)

# Contour plot
plt.contour(X, Y, Z, cmap='viridis')

# Add title and labels
plt.title("Contour Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
````
